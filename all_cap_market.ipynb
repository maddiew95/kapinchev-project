```python
import requests, bs4 as soup, pandas as pd

url = "https://stockanalysis.com/list/"

mega = "mega-cap-stocks/"
large = "large-cap-stocks/"
mid = "mid-cap-stocks/"
small = "small-cap-stocks/"
micro = "micro-cap-stocks/"
nano = "nano-cap-stocks/"
```

```python
r_mega = requests.get(url + mega)
r_large = requests.get(url + large)
r_mid = requests.get(url + mid)
r_small = requests.get(url + small)
r_micro = requests.get(url + micro)
r_nano = requests.get(url + nano)

s_mega = soup.BeautifulSoup(r_mega.text, "html.parser").find_all("tr", class_="svelte-utsffj")[1::] 
s_large = soup.BeautifulSoup(r_large.text, "html.parser").find_all("tr", class_="svelte-utsffj")[1::] 
s_mid = soup.BeautifulSoup(r_mid.text, "html.parser").find_all("tr", class_="svelte-utsffj")[1::] 
s_small = soup.BeautifulSoup(r_small.text, "html.parser").find_all("tr", class_="svelte-utsffj")[1::] 
s_micro = soup.BeautifulSoup(r_micro.text, "html.parser").find_all("tr", class_="svelte-utsffj")[1::] 
s_nano = soup.BeautifulSoup(r_nano.text, "html.parser").find_all("tr", class_="svelte-utsffj")[1::] 

```

```python
s_all = s_mega + s_large + s_mid + s_small + s_micro + s_nano
```

```python
data = []
for i in s_all:
   metadata = []
   for j in i.find_all("td"):
      metadata.append(j.text)
   data.append(metadata)
```

```python
total_markets = pd.DataFrame(data, columns=["n", "Symbol", "Name","Market Cap", "Stock Price", "Change", "Revenue"]).drop(columns=["n"])
```

```python
total_markets["Market Cap"].replace({",": ""})
```

```python
total_markets["Market Cap"] = total_markets["Market Cap"].apply(lambda x: int(float(x.replace("B", ""))* 10**9) if "B" in x else int(float(x.replace("M", ""))* 10**6))
```
